```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from statsforecast import StatsForecast
from statsforecast.models import AutoARIMA
```

```python
df = pd.read_csv('processed-time-series.csv')

cols_to_rename = {"Date": "ds","SubGrup proced":"unique_id","Value": "y"}
col_order = ['unique_id', 'ds', 'y']

df
```

```python
df.rename(columns=cols_to_rename, inplace=True)
df = df[col_order]
print(df)
```

```python
sf = StatsForecast(
    models=[AutoARIMA(season_length=12)],
    freq='M'
)
sf.fit(df)
```

```python
forecast_df = sf.predict(h=12, level=[90])

forecast_df.tail()
```

```python
df["ds"]=pd.to_datetime(df["ds"])
sf.plot(df, forecast_df, level=[90], engine='plotly')
```

```python
df.to_csv('/statsforecast-ts.csv') # Incluindo os dados da Covid
forecast_df.to_csv('/forecasted-df.csv') # Incluindo os dados da COVID
```
